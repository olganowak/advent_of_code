```python
example = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""

example_list = [list(line) for line in example.split()]
```

```python
with open("input.txt") as f:
    list_of_lists = [list(line) for line in f.read().split()]
list_of_lists
```

```python
e = example_list.copy()
counter = 0
for i, line in enumerate(e):
    next_line_index = set()
    if i == 0: 
        line_indexes = {line.index("S")}
    else:
        for line_index in line_indexes:
            if line[line_index] == ".":
                line[line_index] = "|"
                next_line_index.add(line_index)
            elif line[line_index] == "^":
                counter += 1
                if line_index != 0:
                    next_line_index.add(line_index - 1)
                if line_index != len(line) - 1:
                    next_line_index.add(line_index + 1)
    if next_line_index:
        line_indexes = next_line_index
    print(line)

counter
```

```python
e = example_list.copy()
for i, line in enumerate(e):
    next_line_counts = {}
    if i == 0: 
        line_counts = {line.index("S"): 1}
    else:
        for index, count in line_counts.items():
            if line[index] == ".":
                next_line_counts[index] = next_line_counts.get(index - 1, 0) + count
            elif line[index] == "^":
                if index != 0:
                    next_line_counts[index - 1] = next_line_counts.get(index - 1, 0) + count
                if index != len(line) - 1:
                    next_line_counts[index + 1] = next_line_counts.get(index + 1, 0) + count
                    
    # after processing this row, move to the new set of beams
    if next_line_counts: 
        line_counts = next_line_counts
    print(line_counts)
    print(line)

sum(line_counts.values()) 
```
