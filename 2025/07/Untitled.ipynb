```python
example = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""

example_list = [list(line) for line in example.split()]
```

```python
with open("input.txt") as f:
    list_of_lists = [list(line) for line in f.read().split()]
list_of_lists
```

```python
e = [row[:] for row in list_of_lists]
counter = 0
for i, line in enumerate(e):
    next_line_index = set()
    if i == 0: 
        line_indexes = {line.index("S")}
    else:
        for line_index in line_indexes:
            if line[line_index] == ".":
                line[line_index] = "|"
                next_line_index.add(line_index)
            elif line[line_index] == "^":
                counter += 1
                if line_index != 0:
                    next_line_index.add(line_index - 1)
                if line_index != len(line) - 1:
                    next_line_index.add(line_index + 1)
    if next_line_index:
        line_indexes = next_line_index
    print("".join(line))

counter
```

```python
e = [row[:] for row in example_list]

for i, line in enumerate(e):
    next_line_counts = {}

    if i == 0:
        # start with 1 timeline at S
        line_counts = {line.index("S"): 1}
        continue

    for index, count in line_counts.items():
        cell = line[index]
        if cell == ".":
            # timelines go straight down
            next_line_counts[index] = next_line_counts.get(index, 0) + count
        elif cell == "^":
            # splitter: timelines split left/right
            if index != 0:
                next_line_counts[index - 1] = next_line_counts.get(index - 1, 0) + count
            if index != len(line) - 1:
                next_line_counts[index + 1] = next_line_counts.get(index + 1, 0) + count

    if next_line_counts:
        line_counts = next_line_counts
    print(line_counts)

print("FINAL:", sum(line_counts.values()))
```
