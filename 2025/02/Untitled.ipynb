```python
invalid_ids = []
for start, end in id_ranges:
    for num in range(start, end + 1):
            invalid_ids_num = []
            num_str = str(num)
            for i in range(len(num_str)//2):
                part = num_str[:i+1]
                if len(num_str) % len(part) == 0:
                    if part * (len(num_str)//len(part)) == num_str:
                        invalid_ids_num.append(num)
            if invalid_ids_num:
                for invalid_id in set(invalid_ids_num):
                    invalid_ids.append(invalid_id)
```
