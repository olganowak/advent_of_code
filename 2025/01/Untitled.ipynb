```python
position = 50
positions = [position]
for turn in turns:
    direction = turn[0]
    num = int(turn[1:])
    if direction == "L":
        position = (position - num) % 100
        positions.append(position)
    if direction == "R":
        position = (position + num) % 100
        positions.append(position)
```
