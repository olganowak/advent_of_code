```python
import numpy as np
from itertools import combinations

example = """162,817,812
57,618,57
906,360,560
592,479,940
352,342,300
466,668,158
542,29,236
431,825,988
739,650,466
52,470,668
216,146,977
819,987,18
117,168,530
805,96,715
346,949,466
970,615,88
941,993,340
862,61,35
984,92,344
425,690,689"""

points = [np.array([int(x) for x in line.split(",")]) for line in example.split()]
```

```python
with open("input.txt") as f:
    points = [np.array([int(x) for x in line.split(",")]) for line in f.read().split()]
```

```python
pairs = list(combinations(points, 2))
```

```python
distances = {}

for p1, p2 in pairs: 
    distances[np.linalg.norm(p2 - p1)] = [tuple(p1), tuple(p2)]

distances = dict(sorted(distances.items()))
```

```python
circuits = {}
for i, point in enumerate(points):
    circuits[i+1] = [tuple(point)]
```

```python
points = {}
for c, p in circuits.items(): 
    for point in p:
        points[point] = c

points
```

```python
C = 1_000
counter = 0 
for distance, (p1, p2) in distances.items():
    c1 = points[p1]
    c2 = points[p2]
    if c1 != c2:
        for p in circuits[max(c1,c2)]:
            circuits[min(c1, c2)].append(p)
        del circuits[max(c1,c2)]
    points = {}
    for c, p in circuits.items(): 
        for point in p:
            points[point] = c
    counter += 1
    if counter == C:
        break   
```

```python
len_circuits = []
for c, p in circuits.items():
    len_circuits.append(len(p))

len_circuits.sort(reverse=True)
len_circuits
```

```python
result = len_circuits[0] * len_circuits[1] * len_circuits[2]
result
```

```python
def part2(distances, circuits, points):
    for distance, (p1, p2) in distances.items():
        c1 = points[p1]
        c2 = points[p2]
        if c1 != c2:
            for p in circuits[max(c1,c2)]:
                circuits[min(c1, c2)].append(p)
            del circuits[max(c1,c2)]
        points = {}
        for c, p in circuits.items(): 
            for point in p:
                points[point] = c
        if len(circuits) == 1:
            return p1[0] * p2[0]
    
part2(distances, circuits, points)
```
