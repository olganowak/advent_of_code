```python
lines = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   + """.split("\n")
    
operators = lines[-1].split()
numbers_horizontal = [[int(x) for x in line.split()] for line in lines[:-1]]

C = len(numbers_horizontal[0])

numbers_vertical = []
for index in range(C): 
    numbers_vertical.append([row[index] for row in numbers_horizontal])

numbers_vertical
```

```python
def multiply(list_of_nums):
    if len(list_of_nums) == 0:
        return None
    elif len(list_of_nums) == 1:
        return list_of_nums[0]
    else:
        if len(list_of_nums) >= 2:
            result = list_of_nums[0] * list_of_nums[1]
        if len(list_of_nums) >= 3:
            for num in list_of_nums[2:]:
                result *= num
        return result

results = []
for nums, operator in zip(numbers_vertical, operators):
    if operator == "*":
        results.append(multiply(nums))
    elif operator == "+":
        results.append(sum(nums))

sum(results)
```
