```python
with open("input.txt") as f:
    map = [list(line) for line in f.read().split()]

map
```

```python
def fewer_than_4(map, row, col):
    num_of_paper = 0
    # north
    if row != 0:
        if map[row-1][col] == "@":
            num_of_paper += 1
    # northeast
    if row != 0 and col != len(map[row])-1:
        if map[row-1][col+1] == "@":
            num_of_paper += 1
    # east
    if col != len(map[row])-1:
        if map[row][col+1] == "@":
            num_of_paper += 1
    # southeast
    if row != len(map) - 1 and col != len(map[row])-1:
        if map[row+1][col+1] == "@":
            num_of_paper += 1
    # south
    if row != len(map) - 1:
        if map[row+1][col] == "@":
            num_of_paper += 1
    # southwest
    if row != len(map) - 1 and col != 0:
        if map[row+1][col-1] == "@":
            num_of_paper += 1
    # west
    if col != 0:
        if map[row][col-1] == "@":
            num_of_paper += 1
    # northwest
    if row != 0 and col != 0:
        if map[row-1][col-1] == "@":
            num_of_paper += 1
    return num_of_paper < 4

counter = 0
for row, line in enumerate(map):
    for col, element in enumerate(line):
        if element == "@":
            if fewer_than_4(map, row, col):
                counter += 1
print(counter)
```

```python
mega_counter = 0
counter = 1

while counter > 0:
    counter = 0
    for row, line in enumerate(map):
        for col, element in enumerate(line):
            if element == "@":
                if fewer_than_4(map, row, col):
                    counter += 1
                    map[row][col] = "."
    mega_counter += counter

mega_counter
```
