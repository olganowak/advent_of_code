```python
rules_l = example.split('\n\n')[0].split('\n')
rules_l = [[int(x) for x in line.split('|')] for line in rules_l]
rules_l
rules = {}
for rule in rules_l:
    if rule[0] in rules:
        rules[rule[0]].append(rule[1])
    else: 
        rules[rule[0]] = [rule[1]]

rules
```

```python
sorted_rules = dict(sorted(rules.items(), key=lambda item: len(item[1]), reverse=True))
print(sorted_rules)
sorted_pages = [key for key in sorted_rules]
sorted_pages.append(rules[sorted_pages[-1]][0])
sorted_pages
```
